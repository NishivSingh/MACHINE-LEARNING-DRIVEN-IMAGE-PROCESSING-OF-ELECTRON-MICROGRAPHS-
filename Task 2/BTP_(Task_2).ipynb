```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pylab as plt
```

```python
# Load the TEM image
image = cv2.imread('tem_image.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
#Dimension of the image
image.shape
```

```python
# Define the new dimensions (width and height) for the interpolated image
N = 2
new_width = N * image.shape[1]
new_height = N * image.shape[0]
```

```python
# Perform image interpolation
interpolated_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(image, cmap='gray')
ax.axis('off')
plt.show()
```

```python
interpolated_image.shape
```

```python
# Define the parameters for bilateral filtering
diameter = 5     # Reduce the diameter to capture a smaller neighborhood
sigma_color = 50 # Reduce sigma_color for less color space filtering
sigma_space = 50 # Reduce sigma_space for a smaller spatial neighborhood

# Apply bilateral filtering
smoothed_image = cv2.bilateralFilter(image, diameter, sigma_color, sigma_space)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(smoothed_image,cmap='gray')
ax.axis('off')
plt.show()
```
